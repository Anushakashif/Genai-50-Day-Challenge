```python
from google import genai
from google.genai import types
import os
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.environ["GOOGLE_API_KEY"]

client = genai.Client(api_key=API_KEY)
```

### Understanding User Prompts vs System Prompts

### What is User Prompt?

A user prompt is a direct instruction or question you send to AI model. It's what the end user asks for

### What is a System Prompt?

A sytem prompt is a hidden instruction that sets the behavior, tone, and role of AI. It shapes how AI responds t user prompts. 

Exxample 1: User Prompt WITHOUT System

```python
# User prompt only
user_prompt_1 = "Write a short and simple description of a large language model in 2–3 sentences. Explain what an LLM is and what it is used for in everyday AI applications."

response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=user_prompt_1
)

print('WITHOUT SYSTEM PROMPT:')
print(response.text)
```

### Example 2: SAME User Prompt WITH Different System prompt

Notice how the same user prompt produces different outputs based on the system prompt

```python
user_prompt = 'Write a short and simple description of a large language model in 2–3 sentences. Explain what an LLM is and what it is used for in everyday AI applications.'

print('='*80)
print('System prompt 1: You are an Expert AI Assistant')
print('='*80)

response1 = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=user_prompt,
    config=types.GenerateContentConfig(
        system_instruction="""
You are an expert AI assistant explaining large language models.
Follow these rules strictly:
Explain what a large language model is in clear, professional language.
Describe its main purpose and real-world applications.
Use simple but informative technical explanation.
Keep the response between 8 and 12 sentences.
Do not use bullet points or lists in the final answer.
Maintain an educational and neutral tone.
Do not include unnecessary story-style or poetic language.
"""
    )
)

print(response1.text)
```

```python
user_prompt = 'Write a cold outreach email as an AI/ML engineer introducing my services to potential clients.'

# System Prompt 2: Exxpert graphic designer
print('='*80)
print('System Prompt: You are an experienced AI/ML engineer')
response2 = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=user_prompt,
    config=types.GenerateContentConfig(
        system_instruction="You are an experienced AI/ML engineer writing a professional cold outreach email to potential clients. The email should clearly introduce who I am, mention that I specialize in artificial intelligence and machine learning solutions, and briefly explain how I can help businesses solve real-world problems using AI. The tone should be professional, confident, and friendly but not pushy. Keep the email concise (around 120–180 words). Include a strong opening line, a short description of services (such as model development, automation, data analysis, or deployment), and a call to action encouraging the client to reply or schedule a discussion. Avoid technical jargon and focus on business value rather than implementation details. End the email politely with my name as an AI/ML engineer."
    )
)
print(response2.text)
```
