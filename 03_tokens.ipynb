### How Are Tokens Calclated (Using OpenAI'S Token Calculator)

Let's use tiktoken - OpenAI's offline token counter - to see exxactly how text is tokenized

```python
import tiktoken 

encoding = tiktoken.encoding_for_model('gpt-5-nano')

tokens = encoding.encode('Hi, my name is Anusha and I like to read')
```

```python
tokens
```

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} = {token_text}")
```

### What are Tokens?

A token is a small unit of text (a word r part of a word). Every prompt and response usess tokens, which affects:
- Cost: You pay per token used
- Speed: Fewer tokens = faster response
- Quality: Sometimes shorter responses are better, sometimes you need more detail
