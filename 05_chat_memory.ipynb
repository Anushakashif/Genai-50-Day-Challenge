### Illusion of "Memory"

```python
from google import genai
from google.genai import types
import os
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.environ['GOOGLE_API_KEY']
client = genai.Client(api_key=API_KEY)
```

```python
# First message: Introduce yourself
response_1 = client.models.generate_content(
    model='gemini-2.5-flash',
    contents='Hi, I am Anusha. I work as an AI Engineer'
)

print('Assistant:', response_1.text)
```

### OK let's ask a follow-up question

```python
response_2 = client.models.generate_content(
    model='gemini-2.5-flash',
    contents='What is my name and what i do?'
)

print('Assistant', response_2.text)
```

Here's the thing: every call to an LLM is completely STATELESS. it's a totally new call, every single time. As AI Engineer, it's OUR JOB to devise techniques to give the impression that the LLM has a "memory

```python
from google import genai
from google.genai import types

# Build conversation history using types.Content 
messages = [
    types.Content(
        role='user',
        parts=[types.Part(text='Hi, I am Anusha')]
    ),
    types.Content(
        role='model',
        parts=[types.Part(text='Hi, Anusha! How can i assist you today')]
    ),
    types.Content(
        role='user',
        parts=[types.Part(text='What/s my name?')]
    )
]

# Get response
response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=messages
)

print('Assistant:', response.text)
```
